```python
import zipfile
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import splprep, splev

# Extract the uploaded zip file
uploaded_file = 'problems-1.zip'
extracted_dir = '/mnt/data/problems-2'

with zipfile.ZipFile(uploaded_file, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir)

# Function to list files in a directory
print("Listing all files in the extracted directory:")
for root, dirs, files in os.walk(extracted_dir):
    for filename in files:
        print(os.path.join(root, filename))

# Function to read CSV data into structured format
def read_csv(csv_path):
    np_path_XYs = np.genfromtxt(csv_path, delimiter=',')
    path_XYs = []
    for i in np.unique(np_path_XYs[:, 0]):
        npXYs = np_path_XYs[np_path_XYs[:, 0] == i][:, 1:]
        XYs = []
        for j in np.unique(npXYs[:, 0]):
            XY = npXYs[npXYs[:, 0] == j][:, 1:]
            XYs.append(XY)
        path_XYs.append(XYs)
    return path_XYs


# Corrected path to the CSV file containing the curves data
complex_path = os.path.join(extracted_dir, 'problems/isolated.csv')

if os.path.exists(complex_path):
    paths_XYs = read_csv(complex_path)

    # Function to plot the paths with smoothing
    def plot_smoothed(paths_XYs):
        fig, ax = plt.subplots(tight_layout=True, figsize=(8, 8))
        colours = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

        for i, XYs in enumerate(paths_XYs):
            c = colours[i % len(colours)]
            for XY in XYs:
                # Fit a B-spline to the data points
                tck, u = splprep(XY.T, s=0)
                u_new = np.linspace(u.min(), u.max(), 1000)
                x_new, y_new = splev(u_new, tck, der=0)

                ax.plot(x_new, y_new, c=c, linewidth=2)
        ax.set_aspect('equal')
        plt.show()

    plot_smoothed(paths_XYs)
else:
    print(f"File {complex_path} not found.")
```

```python
import zipfile
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import math

# Step 1: Unzip the file and extract the data
uploaded_file = 'problems-1.zip'

with zipfile.ZipFile(uploaded_file, 'r') as zip_ref:
    zip_ref.extractall('/mnt/data/problems-2')

extracted_dir = '/mnt/data/problems-2'

print("Listing all files in the extracted directory:")
for root, dirs, files in os.walk(extracted_dir):
    for filename in files:
        print(os.path.join(root, filename))

# Step 2: Read the CSV file
def read_csv(csv_path):
    np_path_XYs = np.genfromtxt(csv_path, delimiter=',')
    path_XYs = []
    for i in np.unique(np_path_XYs[:, 0]):
        npXYs = np_path_XYs[np_path_XYs[:, 0] == i][:, 1:]
        XYs = []
        for j in np.unique(npXYs[:, 0]):
            XY = npXYs[npXYs[:, 0] == j][:, 1:]
            XYs.append(XY)
        path_XYs.append(XYs)
    return path_XYs

complex_path = os.path.join(extracted_dir, 'problems/isolated.csv')

if os.path.exists(complex_path):
    paths_XYs = read_csv(complex_path)

    # Step 3: Plot the original detected shape
    def plot(paths_XYs, bezier_curves=None):
        fig, ax = plt.subplots(tight_layout=True, figsize=(8, 8))
        colours = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
        for i, XYs in enumerate(paths_XYs):
            c = colours[i % len(colours)]
            for XY in XYs:
                ax.plot(XY[:, 0], XY[:, 1], c=c, linewidth=2)
        if bezier_curves is not None:
            for curve in bezier_curves:
                t_values = np.linspace(0, 1, 100)
                bezier_points = np.array([bezier_curve(t, *curve.flatten()) for t in t_values])
                ax.plot(bezier_points[:, 0], bezier_points[:, 1], 'k-', linewidth=1)
        ax.set_aspect('equal')
        plt.show()

    plot(paths_XYs)

    # Step 4: Define Bezier fitting functions
    def bezier_curve(t, *control_points):
        n = len(control_points) - 1
        points = np.array(control_points).reshape((n + 1, 2))
        binomial_coefficients = np.array([math.comb(n, i) for i in range(n + 1)])
        t_powers = np.array([(1 - t) ** (n - i) * t ** i for i in range(n + 1)])
        return np.dot(binomial_coefficients, (points.T * t_powers).T)


    def objective(control_points, points, n_control_points):
        control_points = control_points.reshape((n_control_points, 2))
        t_values = np.linspace(0, 1, len(points))
        bezier_points = np.array([bezier_curve(t, *control_points.flatten()) for t in t_values])
        return np.sum(np.linalg.norm(bezier_points - points, axis=1))

    def fit_bezier(points, n_control_points):
        if len(points) < n_control_points:
            print(f"Not enough points to fit Bezier curve with {n_control_points} control points.")
            return None

        # Ensure the initial guess has the correct shape
        initial_guess = np.linspace(points[0], points[-1], n_control_points).flatten()

        # Fit Bezier curve using minimize
        result = minimize(objective, initial_guess, args=(points, n_control_points), method='L-BFGS-B')

        if result.success:
            try:
                # Ensure reshaping is done correctly
                control_points = result.x.reshape((n_control_points, 2))
                return control_points
            except ValueError as e:
                print(f"Error reshaping control points: {e}")
                return None
        else:
            print(f"Optimization failed for control points: {result.message}")
            return None

    # Step 5: Fit Bezier curve globally
    bezier_curves = []
    for i, XYs in enumerate(paths_XYs):
        for XY in XYs:
            print(f"Processing path {i+1} with {len(XY)} points")
            try:
                # Determine the shape and adjust the number of control points
                if i == 0:  # Example: Square
                    bezier_curve = fit_bezier(XY, n_control_points=4)
                elif i == 1:  # Example: Circle
                    bezier_curve = fit_bezier(XY, n_control_points=8)
                elif i == 2:  # Example: Star
                    bezier_curve = fit_bezier(XY, n_control_points=12)
                else:
                    bezier_curve = fit_bezier(XY, n_control_points=6)  # Default value

                if bezier_curve is not None:
                    bezier_curves.append(bezier_curve)
            except Exception as e:
                print(f"Error fitting Bezier curve for path: {e}")

    # Step 6: Plot the original and fitted curves
    plot(paths_XYs, bezier_curves)
else:
    print(f"File {complex_path} not found.")
```
